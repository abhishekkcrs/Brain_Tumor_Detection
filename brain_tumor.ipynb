```python
### Abhishek Sharma
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
import os
path=os.listdir('C:\\Users\\Dell Guna\\Desktop\\brain-tumor-detection-master\\brain_tumor\\Training\\')
classes={'no_tumor':0,'pituitary_tumor':1}
```

```python
import cv2
X=[]
Y=[]
for cls in classes:
    pth="C:\\Users\\Dell Guna\\Desktop\\brain-tumor-detection-master\\brain_tumor\\Training\\"+cls
    for j in os.listdir(pth):
        print(j)
        img=cv2.imread(pth+'\\'+j,0) #To read img in 2D ... 0 is used
        img=cv2.resize(img,(200,200))
        X.append(img)
        Y.append(classes[cls])
```

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
```

```python
ann = models.Sequential([
        layers.Flatten(input_shape=(200,200)),
        layers.Dense(3000, activation='relu'),
        layers.Dense(1000, activation='relu'),
        layers.Dense(10, activation='softmax')    
    ])

ann.compile(optimizer='SGD',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

ann.fit(X,Y, epochs=5)
```

```python
Y[5]
```

```python
cnn = models.Sequential([
    layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
cnn.compile(optimizer='SGD',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

cnn.fit(X,Y, epochs=5)
```
